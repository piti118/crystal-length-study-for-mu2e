```python
from math import pi,exp,sqrtimport numpy as npdef gauss(x,mu=0.0,sigma=1.0):    ret = 1.0/sqrt(2*pi)/sigma    ret *= exp(-(x-mu)**2/(2*sigma**2))    return retdef gausscdf(edges,mu,sigma):    cum = 0.0    ret = np.zeros(len(edges))    for i in xrange(1,len(edges)):        width = edges[i]-edges[i-1]        area = 0.5*(gauss(edges[i],mu,sigma)+gauss(edges[i-1],mu,sigma))*width        cum+=area        ret[i]= cum    return retdef normgauss(edges,mu,sigma):    cdf = gausscdf(edges,mu,sigma)    vg = np.vectorize(gauss)    v = vg(edges,mu,sigma)    w = (edges[-1]-edges[0])/len(edges)    v /= (cdf[-1]-cdf[0])/w    return vdef test_gauss():    x = np.linspace(-1,2,1000)    vg = np.vectorize(gauss)    y = vg(x)    plot(x,vg(x,1,2))    plot(x,normgauss(x,1,2))#test_gauss()
```
