```python
import numpy as npfrom collections import defaultdictfrom ROOT import *from math import ceilimport array as arr
```

```python
tf = TFile('../square_11.root')tree = tf.Get('hitmap')
```

```python
class HitMap:    size = 21    def __init__(self):        self.hit = np.zeros((self.size,self.size))    def acc_root(self,data):        row = data.l + floor(self.size/2.)        col = data.k + floor(self.size/2.)        self.hit[row,col]=data.E    def sumE(self):        return np.sum(np.sum(self.hit))
```

```python
dd = defaultdict(HitMap)for data in tree:    dd[(data.eventno,data.runno)].acc_root(data)
```

```python
einfo_tree = tf.Get('eventinfo')
```

```python
einfo_dict = defaultdict(list)for einfo in einfo_tree:    einfo_dict[round(einfo.angle)].append((einfo.eventno,einfo.runno))
```

```python
E_map = defaultdict(lambda: arr.array('d'))left = 10;weirdone = []for angle, keys in einfo_dict.items():    for key in keys:        s= dd[key].sumE()        if(s>104.7 and left>0):            left-=1            print key            weirdone.append(key)        E_map[angle].append(s)
```
