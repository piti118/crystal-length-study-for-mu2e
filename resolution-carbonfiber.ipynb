```python
class Hitmap:    def __init__(self,numrow=21,numcol=21):        self.hmap = np.zeros([numrow,numcol])        self.numrow = numrow        self.numcol = numcol    def acc(self,l,k,E):        i,j = self.lk2ij(l,k)        self.hmap[i,j]+=E    def lk2ij(self, l,k):        return l+self.numcol/2,k+self.numrow/2    def sumE(self,cutoff=None):        if cutoff is not None:            return np.sum(np.sum(self.hmap[self.hmap>cutoff]))        else:            return np.sum(np.sum(self.hmap))
```
